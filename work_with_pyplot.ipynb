Легенда

```python
x = [1, 5, 10, 15, 20]
y = [1, 7, 3, 5, 11]
plt.plot(x, y, label='steel price')
plt.title('Char price', fontsize=15)
plt.xlabel('Day', fontsize=12, color='blue')
plt.ylabel('Price', fontsize=12, color='blue')
plt.legend()
plt.grid(True)
plt.text(15, 4, 'grow up!')
```

Несколько графиков с разными стилями линий

```python
x = np.array([1,5,10,15,20])
y1 = np.array([1,7,3,5,11])
y2 = y1*1.2 + 1
y3 = y2*1.2 + 1
y4 = y3*1.2 + 1
plt.plot(x, y1, '-', x, y2, '--', x, y3, '-.', x, y4, ':')
```

```python
plt.plot(x, y1, '-')
plt.plot(x, y2, '--')
plt.plot(x, y3, '-.')
plt.plot(x, y4, ':')
```

```python
x = [1, 5, 10, 15, 20]
y = [1, 7, 3, 5, 11]
plt.plot(x, y, '--r')
```

```python
plt.plot(x, y, 'bx')
plt.plot(y, x, 'ro')

```

Размещение графиков на разных полях

```python
x = np.array([1, 5, 10, 15, 20])
y1 = np.array([1, 7, 3, 5, 11])
y2 = y1*1.2+1
y3 = y2*1.2+1
y4 = y3*1.2+1
plt.figure(figsize=(12,7))

plt.subplot(2,2,1)
plt.plot(x,y1,'-')
plt.subplot(2,2,3)
plt.plot(x, y2, '--')
plt.subplot(2,2,2)
plt.plot(x, y3, '-.')
plt.subplot(2,2,4)
plt.plot(x,y4,':')
```

Работа с функцией subplots()

```python
fig, axs = plt.subplots(2,2, figsize=(12,7))
axs[0,0].plot(x, y1, '-')
axs[0,1].plot(x, y2, '--')
axs[1,0].plot(x, y3, '-.')
axs[1,1].plot(x, y4, ':')
```
