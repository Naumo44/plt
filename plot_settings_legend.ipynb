```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Отображение легенды

```python
x = [1, 5, 10, 15, 20]
y1 = [1, 7, 3, 5, 11]
y2 = [4, 3, 1, 8, 12]
plt.plot(x, y1, 'o-r', label='line 1')
plt.plot(x, y2, 'o-.g', label='line 2')
plt.legend()
# есть вот такой вариант
# задаём метки для каждого графика, а потом включаем легенду
```

```python
plt.plot(x, y1, 'o-r')
plt.plot(x, y2, 'o-.g')
plt.legend(['L1', 'L2'])
# Здесь мы задаём метки в самой функции отображения легенды в том порядке,
# в котором у нас следуют графики
```

```python
line1, = plt.plot(x, y1, 'o-b')
line2, = plt.plot(x, y2, 'o-.m')
plt.legend((line2, line1), ['L2', 'L1'])
# В этом варианте мы вручную назначаем порядок определения меток
```

Строковое определение расположения меток легенды, тут без комментариев

```python

locs = ['best', 'upper right', 'upper left', 'lower left', 
        'lower right', 'right', 'center left', 'center right', 
        'lower center', 'upper center', 'center']
plt.figure(figsize=(12, 12))
for i in range(3):
    for j in range(4):
        if i*4+j < 11:
            plt.subplot(3, 4, i*4+j+1)
            plt.title(locs[i*4+j])
            plt.plot(x, y1, 'o-r', label='line 1')
            plt.plot(x, y2, 'o-.g', label='line 2')
            plt.legend(loc=locs[i*4+j])
        else:
            break
```

Для более гибкого управления положением легенды есть метод
bbox_to_anchor(x, y, width, height)

```python
plt.plot(x, y1, 'o-r', label='line 1')
plt.plot(x, y2, 'o-.g', label='line 2')
plt.legend(bbox_to_anchor=(0.5, 0.5))
# Тут можно поиграться с координатами, чтобы проще было понять, как это
# работает
```

Доп параметры отображения объекта легенды

```python
plt.plot(x, y1, 'o-r', label='line 1')
plt.plot(x, y2, 'o-.g', label='line 2')
plt.legend(fontsize=14, shadow=True, framealpha=1, facecolor='y', edgecolor='r', title='Легенда')
```
