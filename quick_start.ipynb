Если работать в Jupyter, то для того, чтобы отображать графики рядом с ячейками с кодом, нужно прописать специальную magic команду

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)
%matplotlib inline
```

Основные элементы графика в библиотеке



```python
x = np.linspace(0, 10, 10)
y1 = 4*x
y2 = x**2

fig, ax = plt.subplots(figsize=(8, 6))

ax.set_title("Графики зависимостей: y1=4*x, y2=x^2", fontsize=16)
ax.set_xlabel("x", fontsize=14)        
ax.set_ylabel("y1, y2", fontsize=14)
ax.grid(which="major", linewidth=1.2)
ax.grid(which="minor", linestyle="--", color="gray", linewidth=0.5)
ax.scatter(x, y1, c="red", label="y1 = 4*x")
ax.plot(x, y2, label="y2 = x^2")
ax.legend()
ax.xaxis.set_minor_locator(AutoMinorLocator())
ax.yaxis.set_minor_locator(AutoMinorLocator())
ax.tick_params(which='major', length=10, width=2)
ax.tick_params(which='minor', length=5, width=1)
```

```python
x = np.linspace(0, 10, 50) # от 0 до 10, 15 штук
y = x
```

```python
# построение графика
plt.title("Линейная зависимость y = x") # заголовок
plt.xlabel('x') # подпись оси x
plt.ylabel('y') # подпись оси y
plt.grid() # включение отображения сетки
plt.plot(x, y) # график
```

```python
plt.plot(x, y, 'r--') # тут мы поменяли цвет и сделали линию пунктирной
```

Два графика на одном поле

```python
x = np.linspace(0, 10, 15)
y1 = x
y2 = x**2 # возвели все элементы массива, чтобы отобразить квадратичную функцию

plt.title('Зависимость y1=x, y2=x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.plot(x, y1, x, y2)
```

Несколько разделенных полей с графиками

```python
x = np.linspace(0, 10, 50)
y1 = x
y2 = x**2

plt.figure(figsize=(9,9))

plt.subplot(2, 1, 1)
plt.plot(x, y1)
plt.title('Зависимость: y1=x, y2=x^2')
plt.ylabel('y1', fontsize=14)
plt.grid(True)
plt.subplot(2,1,2)
plt.plot(x, y2)
plt.xlabel('x', fontsize=14)
plt.ylabel('y2', fontsize=14)
plt.grid(True)
```
